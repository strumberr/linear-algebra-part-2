```python

from itertools import permutations
import numpy as np

def permutation_sign(perm):
    inversions = 0
    n = len(perm)
    
    for i in range(n):
        for j in range(i + 1, n):
            if perm[i] > perm[j]:
                inversions += 1
    
    return 1 if inversions % 2 == 0 else -1

def determinant(A):

    n = A.shape[0]
    assert A.shape[0] == A.shape[1]
    
    print(len(A))

    det = 0.0
    elements = []
    sign = 1


    ### YOUR CODE HERE ###
    for perm in permutations(range(n)):
        convert_perms = list(perm)
        
        sign = permutation_sign(perm)
        
        product = 1
        for i in range(n):
            product *= A[i, perm[i]]
            
        det += sign * product
        elements.append(sign * product)
    
    
    return det, sorted(elements)

A = np.array([
    [1, 1, 1],
    [0, 1, 1],
    [0, 0, 1]
])

print(determinant(A))
```
