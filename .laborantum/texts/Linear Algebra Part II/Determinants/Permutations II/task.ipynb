```python

from itertools import permutations

def determinant(A):
    # Ensure the matrix is square
    n = A.shape[0]
    assert A.shape[0] == A.shape[1]
    
    print(len(A))

    det = 0
    elements = []

    ### YOUR CODE HERE ###

    print(A)
    
    for el in permutations(A):
        print(el)
    
    print("\n-----------------\n")

    n = len(A)
    det = 0
    elements = []
    
    for perm in permutations(range(n)):
        product = 1
        inversions = 0
        
        for i in range(n):
            product *= A[i][perm[i]] # here we multiply the elements of the matrix by the permutation

            print(f"Matrix element: {A[i]}")
            print(f"Permutation element: {perm[i]}")
            
            for j in range(i + 1, n):
                if perm[i] > perm[j]:
                    inversions += 1
    
        
        if inversions % 2:
            sign = -1
        else:
            sign = 1
        
        contribution = sign * product
        det += contribution
        elements.append((perm, sign, product, contribution))

    

    return det, sorted(elements)
```
