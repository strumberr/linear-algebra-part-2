```python
import numpy as np
import json_tricks
import scipy
```

```python

def task_1(x, y):
    
    χ = x**2 + y
    ξ = x * y
    
    
    # derivatives
    dχ_dx = 2 * x
    dχ_dy = 1
    dξ_dx = y
    dξ_dy = x
    
    J = np.array(
        [
            [dχ_dx, dχ_dy],
            [dξ_dx, dξ_dy]
        ]
    )
    
    # LU decomposition
    p, l, u = scipy.linalg.lu(J)
    det_l = np.prod(np.diag(l))
    det_u = np.prod(np.diag(u))
    det_J = det_l * det_u

    J = np.array(
        J
    )
    print("J = ", J)
    det_J = (
        det_J
    )
    return J, det_J


def task_2(r, phi):
    
    # I used this: https://byjus.com/maths/jacobian/
    
    print("r = ", r)
    print("phi = ", phi)
    
    x = r * np.sin(phi)
    y = r * np.cos(phi)
    
    # derivatives
    dx_dr = np.sin(phi)
    dx_dphi = r * np.cos(phi)
    dy_dr = np.cos(phi)
    dy_dphi = - r * np.sin(phi)


    J = np.array(
        [
            [dx_dr, dx_dphi],
            [dy_dr, dy_dphi]
        ]
    )
    
    # LU decomposition
    p, l, u = scipy.linalg.lu(J)
    det_l = np.prod(np.diag(l))
    det_u = np.prod(np.diag(u))
    det_J = det_l * det_u
    
    
    det_J = (
        det_J
    )
    return J, det_J
```

```python
inputs = json_tricks.load('inputs.json')

results = {
    'task1': [],
    'task2': []
}

for args in inputs['task1']:
    results['task1'].append(task_1(**args))

for args in inputs['task2']:
    results['task2'].append(task_2(**args))

json_tricks.dump(results, '.answer.json')
```
