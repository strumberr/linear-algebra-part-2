```python
import numpy as np
import json_tricks
import scipy as sp
import sklearn
import lovely_numpy as ln
from PIL import Image
```

```python
mnist = sklearn.datasets.fetch_openml('mnist_784')
X = np.array(mnist.data)
y = np.array(mnist.target).astype(int)
```

```python
print(ln.lovely(X))
print(ln.lovely(y))
```

```python
# Selecting only zeros
X_Zeros = X[y == 0, :]
```

```python
print(ln.lovely(X_Zeros))
```

```python
n_images = 15
results = [X_Zeros[index:n_images**2:n_images].reshape([-1, 28]).astype('uint8') for index in range(n_images)]

results = np.concatenate(results, axis=1)
Image.fromarray(results)
```

```python
import sklearn.feature_extraction
import sklearn.feature_selection

SVD = sklearn.decomposition.TruncatedSVD(n_components=n_images**2)
SVD.fit(X_Zeros)

components = SVD.components_
components = (components - components.min()) / (components.max() - components.min())
components = (components * 255).astype('uint8')
results = [components[index:n_images**2:n_images].reshape([-1, 28]).astype('uint8') for index in range(n_images)]

results = np.concatenate(results, axis=1)
Image.fromarray(results)
```

```python
n_components = 50

coeffs = SVD.components_[:n_components, :] @ X_Zeros.T
decompressed = coeffs.T @ SVD.components_[:n_components, :]

components = decompressed
components = (components - components.min()) / (components.max() - components.min())
components = (components * 255).astype('uint8')
results = [components[index:n_images**2:n_images].reshape([-1, 28]).astype('uint8') for index in range(n_images)]

results = np.concatenate(results, axis=1)
Image.fromarray(results)
```
